```python
#load dataset of portugese student scores
import pandas as pd 
df = pd.read_csv('student-por.csv',sep=";")
```

```python
df['pass']= df.apply(lambda row: 1 if (row ['G1']+row['G2']+row['G3'])>=35 else 0,axis=1)
```

```python
df = pd.get_dummies(df,columns=["sex","school","address","famsize","Pstatus","Mjob","Fjob","reason","guardian","schoolsup","famsup","paid","activities","nursery","higher","internet","romantic"])
```

```python
#shuffle rows
df = df.sample(frac=1)
#split training and testing data 

df_train = df[:500]
df_test = df[500:]

df_test_att = df_test.drop(['pass'],axis=1)
df_test_pass = df_test['pass']

df_train_att = df_train.drop(['pass'],axis=1)
df_train_pass = df_train['pass']

df_pass = df['pass']
df_att = df.drop(['pass'],axis=1)

#number of passing students
import numpy as np 
passed = float(np.sum(df_pass))
total = len(df_pass)

print(f"{passed} students passed out of {total}")
```

```python
#create the decision tree
from sklearn import tree
t = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)
t = t.fit(df_train_att.values,df_train_pass)
```

```python
#view tree score 
t.score(df_test_att.values,df_test_pass)
```

```python
#cross verify results
from sklearn.model_selection import cross_val_score
scores = cross_val_score(t,df_att,df_pass,cv=5)

print("Accuracy:%0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```python
#make prediction
for i in range(10):
    prediction = t.predict([df_test_att.iloc[i].values])
    verify = df_test.iloc[i]['pass']
    print("Prediction:",prediction,
          "\tVerify:",verify,
          "\tCorrect:", str(verify==prediction)
         )
    

```
